```python
import gbcolor
import numpy as np
from matplotlib import pyplot as plt
```

```python
tomori = gbcolor.bangdream.mygo.tomori
anon = gbcolor.bangdream.mygo.anon
soyo = gbcolor.bangdream.mygo.soyo
rikki = gbcolor.bangdream.mygo.rikki
rana = gbcolor.bangdream.mygo.rana
mygo = gbcolor.bangdream.mygo.mygo
```

```python
x = np.linspace(0, 10, 1000)
gaussian = np.exp(-((x - 5) ** 2) / 6)
line_tomori = np.sin(x) * gaussian
line_anon = np.cos(0.5*x) * gaussian
line_soyo = np.sin(3*x + 0.5) * gaussian
line_rikki = np.cos(1.5*x + 0.5) * gaussian
line_rana = np.sin(2*x + 1)  * gaussian
line_mygo = np.zeros(x.shape)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, line_tomori, color=tomori, label='Tomori', linewidth=1)
plt.plot(x, line_anon, color=anon, label='Anon', linewidth=1)
plt.plot(x, line_soyo, color=soyo, label='Soyo', linewidth=1)
plt.plot(x, line_rikki, color=rikki, label='Rikki', linewidth=1)
plt.plot(x, line_rana, color=rana, label='Rana', linewidth=1)
plt.plot(x, line_mygo, color=mygo, label='MyGO', linewidth=1)
plt.title('Mayoi Uta')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('img/Mayoi_Uta.png')
plt.show()
```
